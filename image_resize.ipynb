```python
from PIL import Image
import os
import matplotlib.pyplot as plt

# Function to resize the image
def resize_image(input_image_path, output_image_path, required_size):
    try:
        # Open the input image
        img = Image.open(input_image_path)

        # Display the input image using matplotlib
        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.title("Input Image")
        plt.axis('off')  # Hide axis
        plt.show()

        # Resize the image
        img_resized = img.resize(required_size)

        # Save the resized image
        img_resized.save(output_image_path)
        print(f"Resized image saved at {output_image_path}")

        # Display the resized image using matplotlib
        plt.figure(figsize=(6, 6))
        plt.imshow(img_resized)
        plt.title("Resized Image")
        plt.axis('off')  # Hide axis
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file '{input_image_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_image_path = input('Enter the path of the input image: ')  # Path to your input image
output_image_path = input('Enter the path where the resized image will be saved (e.g., output.jpg): ')  # Dynamic output path
required_size = (32, 32)  # Specify the required size (width, height)

print('The image will be resized to',required_size,'pixels')

# Check if the input image exists
if os.path.exists(input_image_path):
    resize_image(input_image_path, output_image_path, required_size)
else:
    print(f"Error: The file '{input_image_path}' does not exist.")
```

# Upscaling

```python
from PIL import Image
import os
import matplotlib.pyplot as plt

# Function to upscale the image
def upscale_image(input_image_path, output_image_path, upscale_factor):
    try:
        # Open the input image
        img = Image.open(input_image_path)

        # Display the input image using matplotlib
        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.title("Input Image")
        plt.axis('off')  # Hide axis
        plt.show()

        # Calculate the new size for upscaling
        width, height = img.size
        new_size = (int(width * upscale_factor), int(height * upscale_factor))

        # Upscale the image using the LANCZOS filter for better quality
        img_upscaled = img.resize(new_size, Image.LANCZOS)

        # Save the upscaled image
        img_upscaled.save(output_image_path)
        print(f"Upscaled image saved at {output_image_path}")

        # Display the upscaled image using matplotlib
        plt.figure(figsize=(6, 6))
        plt.imshow(img_upscaled)
        plt.title(f"Upscaled Image (Factor {upscale_factor}x)")
        plt.axis('off')  # Hide axis
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file '{input_image_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_image_path = input('Enter the path of the input image: ')  # Path to your input image
output_image_path = input('Enter the path where the upscaled image will be saved (e.g., output_upscaled.jpg): ')  # Dynamic output path
upscale_factor = float(input('Enter the upscale factor (e.g., 2 for 2x upscaling): '))  # Factor by which to upscale the image

# Check if the input image exists
if os.path.exists(input_image_path):
    upscale_image(input_image_path, output_image_path, upscale_factor)
else:
    print(f"Error: The file '{input_image_path}' does not exist.")

```
